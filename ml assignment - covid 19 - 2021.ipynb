```python
d['area_code'] = pd.to_numeric(d['area_code'], errors='coerce')
d['area_code'].fillna(0, inplace=True)
```

```python
d['past_24hrs_symp'] = pd.to_numeric(d['past_24hrs_symp'], errors='coerce')
d['past_24hrs_symp'].fillna(0, inplace=True)
```

```python
d['positve_last14days'] = pd.to_numeric(d['positve_last14days'], errors='coerce')
d['positve_last14days'].fillna(0, inplace=True)
```

```python
d['vaccination'] = pd.to_numeric(d['vaccination'], errors='coerce')
d['vaccination'].fillna(0, inplace=True)
```

```python
d['reason_for_not_vacinated'] = pd.to_numeric(d['reason_for_not_vacinated'], errors='coerce')
d['reason_for_not_vacinated'].fillna(0, inplace=True)
```

```python
d['gender'] = pd.to_numeric(d['gender'], errors='coerce')
d['gender'].fillna(0, inplace=True)
```

```python
d['age'] = pd.to_numeric(d['age'], errors='coerce')
d['age'].fillna(0, inplace=True)
```

```python
d['education'] = pd.to_numeric(d['education'], errors='coerce')
d['education'].fillna(0, inplace=True)
```

```python
d['how_many_got_vacinated_in_family'] = pd.to_numeric(d['how_many_got_vacinated_in_family'], errors='coerce')
d['how_many_got_vacinated_in_family'].fillna(0, inplace=True)
```

```python
d['pregnancy'] = pd.to_numeric(d['pregnancy'], errors='coerce')
d['pregnancy'].fillna(0, inplace=True)
```

```python
plt.hexbin(a['no_of_pepole_in_house'], a['how_many_got_vacinated_in_family'], gridsize=20, cmap='Blues')
plt.xlabel('Number of People in House')
plt.ylabel('Number of People Vaccinated in Family')
plt.title('Hexbin Plot: People in House vs Vaccinated in Family')
plt.colorbar(label='Count in Bin')
plt.show()
```

```python
pregnancy_counts = a['pregnancy'].value_counts()
vaccination_counts = a.groupby('pregnancy')['vaccination'].value_counts().unstack()

fig, ax = plt.subplots()

ax.pie(pregnancy_counts, labels=pregnancy_counts.index, autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.3))
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

ax.legend(vaccination_counts.columns, title='Vaccination', loc='center right', bbox_to_anchor=(1, 0, 0.5, 1))

plt.title('Donut Chart: Vaccination by Pregnancy Status')
plt.show()


#  below chart represents about the pregrent women who are vaccinated 
```

```python
sns.countplot(x='vaccination', hue='pregnancy', data=a, palette='Set2')
plt.title('Stacked Countplot: Vaccination by Pregnancy Status')
plt.show()
```
